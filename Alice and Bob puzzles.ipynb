## Cryptography Problems:


### Corrupt Postal System:

Problem Statement:


In a corrupt postal system each letter is opened and the content
stolen independently of its value. Only securely closed strong
boxes are delivered reliably (because it takes too much hassle to
open them).
How can Bob send a valuable item to Alice in some strong box
which can be locked with several locks when they both can communicate
about the transfer?


Solution:

The problem is to let Alice get at the content of the box. The two
agree on the following procedure:
- 1. Bob sends the box locked by his lock to Alice.
- 2. Alice additionally locks the box she received by her lock and
sends it back to Bob.
- 3. Bob removes his lock from the box and sends the box locked by
only Alice's lock back to Alice.











### Equal Opportunity:
Problem Statement:

Alice and Bob live in different cities and decide to go to see each
other in turns. They want to finnd out who starts to drive to the
other by tossing a coin(ie based on chance).
How do they and out if they live in different cities?

Solution:


Suppose they agree say per e-mail on the following procedure:
- 1. Alice and Bob agree to use a suitable one way function f, so that y=f(x) is easily and f_inverse(y) is    extremely hard to compute.
- 2. Now, say Alice starts and chooses an odd or even x. Now she sends y with y = f(x) to Bob without specifying x.
- 3. Bob receives y and bets whether x was odd or even.
   (If he wins then Alice otherwise Bob has to drive.)
- 4. Alice checks Bob's bet and sends x to Bob for verication, i.e.
to let Bob compare f(x) with the y he initially received.
